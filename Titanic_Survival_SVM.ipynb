```python
import pandas as pd
data = pd.read_csv("titanic.csv")
data.head()
```

```python
sex = pd.get_dummies(data.Sex)
```

```python
sex.head(2)
```

```python
data['Title'] = data.Name.str.split(r'\s*,\s*|\s*\.\s*').str[0]
```

```python
data.head(5)
```

```python
title = pd.get_dummies(data.Title)
```

```python
title.head(5)
```

```python
data = pd.concat([data,sex,title], axis = 1)
```

```python
data.head(5)
```

```python
data.pop('Name')
```

```python
data.pop('Sex')
```

```python
data.head(5)
```

```python
data.pop('Title')
```

```python

x = data.copy()
y = x.pop('Survived')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.1, random_state = 1)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', gamma=15.0, C=20.0, random_state=1)
```

```python
svm.fit(x_train,y_train)
```

```python
y_pred = svm.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
